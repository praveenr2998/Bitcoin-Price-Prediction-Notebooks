```python
# Converting the Timestamp column from string to datetime
historic_data['Timestamp'] = [datetime.fromtimestamp(x) for x in historic_data['Timestamp']]
```

```python
# Function to handel missing values
def fill_missing(df):
    df['Open'] = df['Open'].interpolate()
    df['Close'] = df['Close'].interpolate()
    df['High'] = df['High'].interpolate()
    df['Low'] = df['Low'].interpolate()
    df['Weighted_Price'] = df['Weighted_Price'].interpolate()
    df['Volume_(BTC)'] = df['Volume_(BTC)'].interpolate()
    df['Volume_(Currency)'] = df['Volume_(Currency)'].interpolate()


    print(df.head())
    print("\n")
    print(df.isnull().sum())
```

```python
fill_missing(historic_data)
```

```python


# Setting timestamp as index
historic_data_non_indexed = historic_data.copy()
historic_data = historic_data.set_index('Timestamp')
historic_data.head(10)
```

```python
#Resampling data to daily values
historic_data_daily = historic_data.resample("24H").mean() 
```

```python


import pandas as pd
from datetime import datetime

historic_data_daily = pd.read_csv("/home/praveen/Desktop/Projects/technocolab/Deployment/Deployement/historic_data_final.csv")

yr = 2011
week = 1
iso_week = []
day=0
tmp_yr=0
for index, row in historic_data_daily.iterrows():
    day=int(row[0][-2:]) 
    tmp_yr=int(row[0][:4])

    if tmp_yr == yr and day%7==1:
        week+=1
        ans = str(tmp_yr)+"-"+str(week)
        iso_week.append(ans)
            
    elif tmp_yr == yr and day%7!=1:
        
        ans = str(tmp_yr)+"-"+str(week)
        iso_week.append(ans)

    elif tmp_yr!=yr:
        yr = tmp_yr
        week=1
        ans = str(tmp_yr)+"-"+str(week)
        iso_week.append(ans)
              
              
    

new_iso_week = []
for i in iso_week:
    if i[-2]=="-":
        new_iso_week.append(i[:-1]+"0"+i[-1])
    else:
        new_iso_week.append(i)
    

historic_data_daily['iso_week'] = new_iso_week
```
